```python
from qiskit.primitives import Sampler, Estimator, BackendSampler
from qiskit.algorithms.minimum_eigensolvers import QAOA
from qiskit.algorithms.optimizers import COBYLA, SLSQP, L_BFGS_B, ADAM, SPSA
import matplotlib.pyplot as plt

from utility import Q_to_paulis, symmetric_matrix_generator

maxiter = 200
n_layers = 3
Q = symmetric_matrix_generator(4)
op, _ = Q_to_paulis(Q)
sampler = Sampler()
```

```python
mean_list = []
def QAOA_callback(n_counts, params, mean, std):
    mean_list.append(mean)
```

# COBYLA

```python
optimizer = COBYLA(maxiter=maxiter)
mean_list = []
qaoa = QAOA(
    sampler=sampler,
    optimizer=optimizer,
    reps=n_layers,
    callback=QAOA_callback,
)
result = qaoa.compute_minimum_eigenvalue(op)
p = plt.plot(mean_list)
```
